```python
import spacy

# Lade das deutsche NLP-Modell
nlp = spacy.load("de_core_news_sm")

text = """
Michael stellt sich vor.
Hannah fragt, ob eine Personalbefragung gewünscht ist. Dies ist nicht der Fall.
Hannah eröffnet den Wahlgang.
Nach Rückfrage, ob es noch unabgegebene Stimmzettel gebe, schließt Hannah den Wahlgang.
Wahlergebnis:
ja: 37
nein: 4
Enthaltungen: 1
ungültig: 0
Damit ist Michael Kreuzer als Stellvertreter gewählt. Er nimmt die Wahl an.
"""

# Verarbeite den Text mit spaCy
doc = nlp(text)

# Extrahiere benannte Entitäten
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
import pdfplumber

def extrahiere_segmente(pdf_datei):
    segmente = []
    aktuelles_segment = []
    aktuelle_überschrift = None

    with pdfplumber.open(pdf_datei) as pdf:
        for seite in pdf.pages:
            words = seite.extract_words(extra_attrs=["fontname", "size"])
            for word in words:
                if 'Bold' in word['fontname'] or word['size'] > 10:
                    # Wenn wir auf eine neue fette Überschrift stoßen, speichern wir das bisherige Segment
                    if aktuelle_überschrift:
                        segmente.append({
                            'überschrift': aktuelle_überschrift,
                            'inhalt': ' '.join(aktuelles_segment)
                        })
                    # Setze die aktuelle Überschrift und leere das aktuelle Segment
                    aktuelle_überschrift = word['text']
                    aktuelles_segment = []
                else:
                    aktuelles_segment.append(word['text'])
                    
            # Nach der letzten Seite, füge das letzte Segment hinzu
            if aktuelle_überschrift and aktuelles_segment:
                segmente.append({
                    'überschrift': aktuelle_überschrift,
                    'inhalt': ' '.join(aktuelles_segment)
                })

    return segmente

# Verwendung des Codes
pdf_datei = './data/raw/2021_07_15/protokoll.pdf'
segmente = extrahiere_segmente(pdf_datei)

for segment in segmente:
    print(f"Überschrift: {segment['überschrift']}")
    print(f"Inhalt: {segment['inhalt']}\n")
```
